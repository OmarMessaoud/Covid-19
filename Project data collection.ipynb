```python
# convert date of dose to datetime to calculate the difference
df_death['date_positive'] = pd.to_datetime(df_death['date_positive'])
df_death['date_dose1'] = pd.to_datetime(df_death['date_dose1'])
df_death['date_dose2'] = pd.to_datetime(df_death['date_dose2'])
df_death['date_dose3'] = pd.to_datetime(df_death['date_dose3'])
```

```python
# calculate the difference between vaccine date and positive result date
df_death['days_dose1'] = df_death['date_positive'] - df_death['date_dose1']
df_death['days_dose2'] = df_death['date_positive'] - df_death['date_dose2']
df_death['days_dose3'] = df_death['date_positive'] - df_death['date_dose3']
```

```python
df_death['days_dose1'] = df_death['days_dose1'].fillna(pd.Timedelta(seconds=0))
df_death['days_dose2'] = df_death['days_dose2'].fillna(pd.Timedelta(seconds=0))
df_death['days_dose3'] = df_death['days_dose3'].fillna(pd.Timedelta(seconds=0))
```

```python
# convert type to numeric
from datetime import datetime
df_death['days_dose1']= df_death['days_dose1'].dt.days.astype('int64')
df_death['days_dose2']= df_death['days_dose2'].dt.days.astype('int64')
df_death['days_dose3']= df_death['days_dose3'].dt.days.astype('int64')
```

```python
import numpy as np
df_death['days_dose1']= df_death['days_dose1'].astype(np.float64)
df_death['days_dose2']= df_death['days_dose2'].astype(np.float64)
df_death['days_dose3']= df_death['days_dose3'].astype(np.float64)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
import category_encoders as ce

data_encoded=pd.get_dummies(data=data1,drop_first=True)
data_encoded
```

```python
data_encoded = data_encoded.drop('date_positive', axis = 1)
```

```python
data_encoded.info()
```

```python

X = data_encoded.drop('death',axis = 1)
y = data_encoded.death
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(X_train,y_train)

preds = xg_reg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
params = {"objective":"reg:linear",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=123)
xg_reg = xgb.train(params=params, dtrain=data_dmatrix, num_boost_round=10)
```

```python
import matplotlib.pyplot as plt

xgb.plot_tree(xg_reg,num_trees=0)
plt.rcParams['figure.figsize'] = [50, 10]
plt.show()
```
